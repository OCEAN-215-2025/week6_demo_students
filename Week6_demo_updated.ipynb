## Background

Prochlorococcus is a group of very small phytoplankton; it's average diameter is ~0.5 µm- for reference, the average diameter of human hair is about 50µm!. Despite its small size, this cyanobacteria plays a very important role in ocean ecosystems. Prochlorococcus dominates oligotrophic gyres and can be found virtually anywhere except in the high latitudes. Thus, it plays an important role in biogeochemical cycling, trophic interactions, and global primary production.

## Data manipulation

In this dataset, abundance (`abundance`) is measured in cells/μL, while carbon quota (`Qc_hour`) is measured in pg C/cell. Multiplying these two together, we can compute total carbon contributed by prochlorococcus

```python
df["total_C"] = df["abundance"] * df["Qc_hour"]
display(df)
```

One can also apply numpy mapping function on columns

```python
df["ln_total_C"] = np.log(df["total_C"])
display(df)
```

## Plotting

### Line plot

```python
# plot time versus cell abundance

x = df.index
y = df['abundance']

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot()

ax.plot(x, y, marker='o')

plt.show(fig)
```

```python
import matplotlib.dates as mdates
```

```python
# improve the horizontal axis

x = df.index
y = df['abundance']

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot()

ax.plot(x, y, marker='.')

ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

plt.show(fig)
```

```python
# improve the horizontal axis
# alternative: rotates the tick labels

x = df.index
y = df['abundance']

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot()

ax.plot(x, y, marker='.')

ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))

ax.tick_params(axis="x", labelrotation=45)

plt.show(fig)
```

### Scatter plot

Let's plot latitude against cell abundance

```python
x = df['lat']
y = df['abundance']

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot()

ax.scatter(x, y, marker='.', c='g')
ax.set_xlabel('Latitude (°N)', fontsize=14)
ax.set_ylabel('Abundance (cells/μL)', fontsize=14)
ax.set_title('Latitude and Prochlorococcus Cell Abundance', fontsize=16)
ax.grid()

plt.show(fig)
```
